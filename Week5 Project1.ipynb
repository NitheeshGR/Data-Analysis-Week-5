                                             Data Cleaning

```python
# Converted all columns name in Upper case
df_movie.columns=df_movie.columns.str.upper()
# Droped the rows that contain categorical Nan values
df_movie=df_movie.dropna()
# Changed the name for better understanding
df_movie=df_movie.rename(columns={'DATE_X':'DATE','BUDGET_X':'BUDGET'})
# Checked wheather there is a Null value
df_movie.isnull().sum()
# To check weather the Score contain any categorical data
df_movie.SCORE.unique()

# /converting Budget Data type to integer
df_movie['BUDGET']=df_movie['BUDGET'].astype(int)
# rESET THE iNDEX OF dATAFRAME
df_movie.reset_index(drop=True,inplace=True)
```

```python
df_movie
```

                                TYPE CONVERSION (to_datetime)

```python
# To convert to datetime
df_movie['DATE']=pd.to_datetime(df_movie['DATE'])
# finding unique value in Date
df_movie['DATE'].unique()
# Extracting Year from Date
df_movie['Years']=df_movie['DATE'].dt.year
# To find unique values in years
df_movie['Years'].unique()
```

                                 Groupby Year
                                 

```python
group=df_movie.groupby('Years')['NAMES'].count()
group=group.reset_index(name='Count')
group
```

                               Filtering data for visualisation

```python

gr_1930=group.loc[group['Years']<=1930]['Count'].sum()
gr_1950=group.loc[(group['Years']>1930)&(group['Years']<=1950)]['Count'].sum()
gr_1970=group.loc[(group['Years']>1950)&(group['Years']<=1970)]['Count'].sum()
gr_1990=group.loc[(group['Years']>1970)&(group['Years']<=1990)]['Count'].sum()
gr_2010=group.loc[(group['Years']>1990)&(group['Years']<=2010)]['Count'].sum()
gr_2024=group.loc[(group['Years']>2010)&(group['Years']<=2024)]['Count'].sum()
# Seting the years for visualisation
Year_List=['1903-1930','1931-1950','1951-1970','1971-1990','1991-2010','2011-2023']
# combining all filtered year in one list
group_count_list=[gr_1930,gr_1950,gr_1970,gr_1990,gr_2010,gr_2024]
```

                                      DATA VISUALISATION

```python
plt.plot(Year_List,group_count_list,color='red')
plt.xlabel('Years')
plt.ylabel('Total Movies')
plt.title('Movie Analysis in Years')
plt.show()
```

                                            Movie with high Budget

```python
df_Budget=df_movie.groupby('NAMES')['BUDGET'].max()
df_Budget=df_Budget.reset_index(name='BUDGET')
High_Budget=df_Budget.nlargest(1,'BUDGET')
High_Budget.reset_index(drop=True,inplace=True)
High_Budget
```

                                             Top 5 Scored Film
                                             

```python
# Used to group Data and get Series 
df_Score=df_movie.groupby("NAMES")['SCORE'].max()
df_Score=df_Score.reset_index(name='Score')
Top5=df_Score.nlargest(5,'Score')
Top5=Top5.reset_index(drop=True)
Top5
```
